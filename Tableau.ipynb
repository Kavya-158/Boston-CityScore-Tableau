Check the percentage of null values present in the columns to decide as to keep or delete the data. The day score data has about 30% of null values. Therefore we can drop the rows with null values and continue with the rest of the data.

```python
df_cleaned = df.dropna()
df_cleaned
```

```python
df_cleaned['metric_name'].value_counts()
```

```python
# Create a mapping dictionary for sector names
sector_mapping = {
    'BFD RESPONSE TIME': 'Public safety and Health',
    'EMS RESPONSE TIME': 'Public safety and Health',
    'SIGNAL REPAIR ON-TIME %': 'Transportation Management',
    'STREETLIGHT ON-TIME %': 'Transportation Management',
    'POTHOLE ON-TIME %': 'Transportation Management',
    'SIGN INSTALLATION ON-TIME %': 'Transportation Management',
    'BPS ATTENDANCE': 'Basic city services',
    'CODE ENFORCEMENT ON-TIME %': 'Basic city services',
    'CODE ENFORCEMENT TRASH COLLECTION': 'Basic city services',
    '311 CALL CENTER PERFORMANCE': 'Basic city services',
    'PARKS MAINTENANCE ON-TIME %': 'Basic city services',
    'MISSED TRASH ON-TIME %': 'Basic city services',
    'TREE MAINTENANCE ON-TIME %': 'Basic city services',
    'CITY SERVICES SATISFACTION SURVEYS': 'Constituent satisfaction',
    'GRAFFITI ON-TIME %': 'Basic city services',
    'ON-TIME PERMIT REVIEWS': 'Constituent satisfaction',
    '311 CONSTITUENT EXPERIENCE SURVEYS': 'Constituent satisfaction'
}

# Map the sector names to a new column in the DataFrame
df_cleaned['sector_name'] = df_cleaned['metric_name'].map(sector_mapping)

# Display the updated DataFrame
print(df_cleaned)
```

```python
df_cleaned
```

After the null values are removed we can convert the score values of metrics into binary values for easy understanding as we know that a value above 1 indicates a good day and a value less that 1 is an unsatisfactory day.

```python
columns_to_modify = ['day_score', 'week_score', 'month_score','quarter_score']  # list of column names
df_cleaned[columns_to_modify] = df_cleaned[columns_to_modify].apply(lambda x: np.where(x < 1, 0, 1))
```
